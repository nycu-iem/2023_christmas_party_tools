```python
import pandas as pd

data = pd.read_csv("input.csv", converters={'手機號碼': str})

print(data.columns)

columns_to_keep = ['組別(ex. XX組)','職位 (ex. 組長、工人)','姓名','手機號碼','LINE 暱稱','身分證字號','生日(以民國表示) ex: 92/01/01']
mp = {
    '行政組':1,
    '公關組':2,
    '行銷組':3,
    '活動組':4,
    '場器組':5
}

extended_mp = mp.copy()
extended_mp.update({
    '行政':1,
    '公關':2,
    '行銷':3,
    '活動':4,
    '場器':5
})

print(extended_mp)

for index, row in data.iterrows():
    data.at[index, '組別(ex. XX組)'] = extended_mp[row['組別(ex. XX組)']]

new_data = data[columns_to_keep].rename(columns={
    '職位 (ex. 組長、工人)': '職位',
    '組別(ex. XX組)':'組別',
    '生日(以民國表示) ex: 92/01/01': '生日'
}).sort_values(by=['組別','職位'],ascending=[True,False])

mp = dict(zip(mp.values(), mp.keys()))

for index, row in new_data.iterrows():
    new_data.at[index, '組別'] = mp[row['組別']]

new_data.to_csv("output.csv", index=False, encoding='utf_8_sig')
```
